```python
import pandas as pd
from openai import OpenAI
```

```python
with open('OpenAI - jasperlee3-c.txt', 'r') as infile:
  openai_api_key = infile.read()
client = OpenAI(api_key=openai_api_key)
```

```python
df = pd.read_csv('desc_xyz_bullets.csv', encoding="ISO-8859-1")
df.head()
```

```python
des = [df.des_bullet[0],
       df.des_bullet[4],
       df.des_bullet[3],
       df.des_bullet[7]]

xyz = [df.xyz_bullet[0],
       df.xyz_bullet[4],
       df.xyz_bullet[3],
       df.xyz_bullet[7]]
```

```python
user_input = "{user_input}"

prompt = f"""Remove all the quantifiable achievement. Change the achievement based sentence into description based. You may refer the examples.
input: {xyz[0]}
output: {des[0]}

input: {xyz[1]}
output: {des[1]}

input: {xyz[2]}
output: {des[2]}

input: {xyz[3]}
output: {des[3]}

input: {user_input}
output: """
```

```python
print(prompt)
```

```python
output = []

for bullet in df.loc[df["des_bullet"].isna(), "xyz_bullet"]:
  session = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{
        "role": "user",
        "content": prompt.format(user_input=bullet)
    }],
    temperature=1
  )

  response = session.choices[0].message.content

  print(response)
  output.append(response)
```

```python
df.loc[df["des_bullet"].isna(), "des_bullet"] = output
```

```python
df.to_csv('desc_xyz_bullets_sync_gpt35.csv', index=None)
```
